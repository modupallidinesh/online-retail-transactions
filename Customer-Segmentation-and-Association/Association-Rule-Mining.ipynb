<h1> Association Rule Mining </h1>
<hr>
<h3>In this section, we want to find the items that were often bought together by finding the association rules among them</h3>
<br><br>
In order to use apriori and then association rules(from the mlxtend library), we need to convert the data into standard format with one-hot encoding in which we will group the data by transactions and the name of the products and then unstack it so in each row we would have a transaction and each column will show if a specific product exists in that transaction.

After we run the code, we can see the items that were most often bought together in the table below:

```python
#importing necessary libraries
import pandas as pd, numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

#importing the dataset
CleanDataset = r'../Cleaned-Dataset/OnlineRetail_Cleaned.csv'
Data_Cleaned = pd.read_csv(CleanDataset, index_col = 'InvoiceDate')
Data_Cleaned.index = pd.to_datetime(Data_Cleaned.index, format = '%Y-%m-%d %H:%M', box = False)

#converting the data into the standard form
Baskets = Data_Cleaned.loc[(Data_Cleaned['Quantity']>0) ,['InvoiceNo','Description','Quantity']]
Baskets = Baskets.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack(fill_value=0)
Baskets = (Baskets > 0)

#finding frequent itemsets and association rules
frequent_itemsets = apriori(Baskets, min_support=0.0325, use_colnames=True)
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold=0.5)
rules
```
